```python
import pandas as pd
import matplotlib.pyplot as plt

read_df = pd.read_csv("allscraped.csv")
```

```python
# Plotting
plt.figure(figsize=(8, 5))
plt.scatter( read_df['BlocksRecoveredPct'], read_df['Blocks'], color='skyblue')
plt.axhline(y=65, color="red", linestyle="--") # 65 game mark to indicate over roughly 1 block per game 

plt.axvline(x=read_df['BlocksRecoveredPct'].mean(), color="green", linestyle="--") # Mean of all data
plt.axvline(x=read_df[read_df["Blocks"] > 65]["BlocksRecoveredPct"].mean(), color="blue", linestyle="--") # Mean of more significant data

# Print out general data and significant data means and standard deviations
# Appears to indicate all players average 55% +- 20% and better blockers average 60% +- 5%
# The data also appears to follow a normal distribution
# Still need to check for a specific correlation that can be used with a constant to scale recovery rate by total blocks
print(read_df["BlocksRecoveredPct"].std())
print(read_df["BlocksRecoveredPct"].mean())
print(read_df[read_df["Blocks"] > 65]["BlocksRecoveredPct"].std())
print(read_df[read_df["Blocks"] > 65]["BlocksRecoveredPct"].mean())

plt.show
```
